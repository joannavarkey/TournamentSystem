```python
class Team:
    def __init__(self):
        self.teams = {}  

    def match_record_details(self):
        try:
            tid = int(input("Enter Team ID: "))
            if tid in self.teams:
                raise RuntimeError("Team ID already exists.")

            team_name = input("Enter Team Name: ")
            captain_name = input("Enter Captain Name: ")
            number_batsman = int(input("Enter number of batsman: "))
            number_bowler = int(input("Enter number of bowler: "))
            number_all_rounder = int(input("Enter number of All rounders: "))

            self.teams[tid] = {
                "name": team_name,
                "captain_name": captain_name,
                "batsman": number_batsman,
                "bowlers": number_bowler,
                "all_rounders": number_all_rounder
            }
            print("Team registered successfully.")

        except ValueError:
            print("Invalid input. Counts must be numeric.")
        except RuntimeError as e:
            print(e)
        except Exception as e:
            print("Error:", e)

class TournamentSystem(Team):
    def __init__(self):
        super().__init__()

    def update_team(self):
        try:
            tid = int(input("Enter Team ID to update: "))
            if tid not in self.teams:
                print("Team ID not found.")
                return

            new_name = input("Enter new Team Name: ")
            self.teams[tid]["name"] = new_name
            print("Team information updated.")

        except ValueError:
            print("Invalid ID format.")

    def search_team(self):
        try:
            tid = int(input("Enter Team ID to search: "))
            if tid in self.teams:
                team = self.teams[tid]
                print(f"\n--- Team Details ---")
                print(f"ID: {tid}")
                print(f"Name: {team['name']}")
                print(f"Captain: {team['captain_name']}")
            else:
                print("Team not found.")
        except ValueError:
            print("Invalid ID.")

    def display_team(self):
        if not self.teams:
            print("No records found.")
            return

        print("\n All Team Records")
        for tid, t in self.teams.items():
            print(f"ID: {tid} | Name: {t['name']} | Captain: {t['captain_name']}")

    def menu(self):
        while True:
            print("\n--- Team Management System ---")
            print("1. Add Team")
            print("2. Update Team Name")
            print("3. Search Team")
            print("4. Display All Teams")
            print("5. Exit")

            try:
                choice = int(input("Enter your choice: "))

                if choice == 1:
                    self.match_record_details()
                elif choice == 2:
                    self.update_team()
                elif choice == 3:
                    self.search_team()
                elif choice == 4:
                    self.display_team()
                elif choice == 5:
                    print("Exiting system...")
                    break
                else:
                    print("Invalid choice.")
            except ValueError:
                print("Please enter a valid number.")

if __name__ == "__main__":
    ts = TournamentSystem()
    ts.menu()
```
